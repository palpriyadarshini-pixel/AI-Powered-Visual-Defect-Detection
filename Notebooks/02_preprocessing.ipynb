## Import Required Libraries

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## Load Sample Image

```python
train_path = "../Dataset/train/images"

classes = os.listdir(train_path)

sample_image = os.path.join(
    train_path,
    classes[0],
    os.listdir(os.path.join(train_path, classes[0]))[0]
)

img = cv2.imread(sample_image)

print("Image Shape:", img.shape)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

## Converting Image to NumPy Array

```python
image_array = np.array(img)

print("Data Type:", image_array.dtype)
print("Shape:", image_array.shape)

image_array
```

## Pixel Value Analysis

```python
print("Minimum Pixel Value:", image_array.min())
print("Maximum Pixel Value:", image_array.max())
print("Mean Pixel Value:", image_array.mean())
```

## Image Normalization

```python
normalized_img = image_array / 255.0

print("Minimum Pixel Value:", normalized_img.min())
print("Maximum Pixel Value:", normalized_img.max())
```

## Comparing Original and Normalized Images

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(cv2.cvtColor(image_array, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(normalized_img)
ax[1].set_title("Normalized Image")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```

## Key Observations

- Images are stored as NumPy arrays.
- Pixel values originally range between 0 and 255.
- Normalization scales pixel values between 0 and 1.
- Normalized images are more suitable for neural network training.
