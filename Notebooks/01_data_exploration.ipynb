## Objective

The objective of this notebook is to explore and understand the NEU Surface Defect Dataset before building a deep learning model for visual defect classification.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

## Load Dataset

```python
train_path = "../Dataset/train/images"
val_path = "../Dataset/validation/images"

print("Train Path:", train_path)
print("Validation Path:", val_path)
```

## Explore Defect Categories

```python
classes = os.listdir(train_path)

print("Defect Classes:")
print(classes)
```

## Dataset Statistics

```python
data = []

for defect in classes:
    folder = os.path.join(train_path, defect)

    image_count = len(os.listdir(folder))

    data.append({
        "Defect Type": defect,
        "Number of Images": image_count
    })

df = pd.DataFrame(data)

df
```

## Total Number of Images

```python
total_images = df["Number of Images"].sum()

print("Total Training Images:", total_images)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["Defect Type"],
    df["Number of Images"]
)

plt.title("Distribution of Defect Classes")
plt.xlabel("Defect Type")
plt.ylabel("Number of Images")

plt.xticks(rotation=30)

plt.show()
```

## Sample Images from Each Defect Class

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, defect in enumerate(classes):
    folder = os.path.join(train_path, defect)

    image_name = os.listdir(folder)[0]

    image_path = os.path.join(folder, image_name)

    img = Image.open(image_path)

    axes[i // 3, i % 3].imshow(img, cmap="gray")
    axes[i // 3, i % 3].set_title(defect)
    axes[i // 3, i % 3].axis("off")

plt.tight_layout()
plt.show()
```

## Image Properties

```python
sample_image = os.path.join(
    train_path,
    classes[0],
    os.listdir(os.path.join(train_path, classes[0]))[0]
)

img = Image.open(sample_image)

print("Image Size:", img.size)
print("Image Mode:", img.mode)
```

## Key Observations

- The dataset contains six defect categories.
- Images are grayscale.
- The dataset is already separated into training and validation sets.
- Class distribution is balanced.
- Image dimensions are consistent across the dataset.
- The dataset is suitable for image classification using deep learning models.
